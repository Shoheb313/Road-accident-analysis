```python
# Importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Importing the dataset
url = 'https://raw.githubusercontent.com/damianoc90/Road-accidents-analysis/master/dataset.csv'
dataset = pd.read_csv(url)

# Displaying the first five rows
print(dataset.head())

# Convert 'Accident Date' column to datetime format
dataset['Accident Date'] = pd.to_datetime(dataset['Accident Date'], format='%d/%m/%Y')

# Group by 'Accident Date' and count the number of accidents for each date
accidents_by_date = dataset.groupby('Accident Date').size()

# Plotting
plt.figure(figsize=(12, 6))
accidents_by_date.plot()
plt.title('Number of Accidents Over Time')
plt.xlabel('Accident Date')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.show()


accidents_by_surface = dataset.groupby('Road Surface')['Reference Number'].count().reset_index()

# Plotting
plt.figure(figsize=(10, 6))
plt.bar(accidents_by_surface['Road Surface'], accidents_by_surface['Reference Number'], color='skyblue')
plt.title('Number of Accidents by Road Surface')
plt.xlabel('Road Surface')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)  # Rotate x-axis labels for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# File path with double backslashes
file_path = 'C:\\Users\\pc\\Documents\\Data for Project.csv'

# Or file path with 'r' prefix to indicate a raw string
# file_path = r'C:\Users\pc\Documents\Camera.csv'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path)
print(df.head())

df['Date'] = pd.to_datetime(df['Date'])

# Extract year and month from the 'Date' column
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

# Group the data by year and month and count the number of accidents
monthly_accidents = df.groupby(['Year', 'Month']).size().reset_index(name='Accident Count')

# Plot the number of accidents over time
plt.figure(figsize=(10, 6))
plt.plot(monthly_accidents['Date'], monthly_accidents['Accident Count'], marker='o', linestyle='-')
plt.title('Number of Accidents Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample project data
data = {
    'Task': ['Data collection', 'Data analysis', 'Designing',],
    'Start Date': ['2024-04-05', '2024-04-06', '2024-04-08',],
    'End Date': ['2024-04-06', '2024-04-08', '2024-04-10',]
}

# Convert to DataFrame
df = pd.DataFrame(data)
df['Start Date'] = pd.to_datetime(df['Start Date'])
df['End Date'] = pd.to_datetime(df['End Date'])

# Sort by Start Date
df = df.sort_values(by='Start Date')

# Plotting
plt.figure(figsize=(10, 6))
for i, task in enumerate(df['Task']):
    plt.barh(task, left=df.iloc[i]['Start Date'], width=df.iloc[i]['End Date'] - df.iloc[i]['Start Date'])

# Customizing the plot
plt.xlabel('Timeline')
plt.ylabel('Tasks')
plt.title('Project Gantt Chart')
plt.grid(axis='x')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
